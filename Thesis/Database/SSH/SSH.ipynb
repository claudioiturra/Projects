```matlab
clear all,close all;
load SSH.mat;
load('/home/cl4/Insync/claudioiturra@oceanografia.udec.cl/Google Drive/InSync/temp/wind/windstations/slabModelX/pwp/chileMAP.mat');
idx = datevec(timefix);
ind = find(idx(:,1) == 2005 & idx(:,2) == 1);
```

```matlab
for i = 1:4;
subplot(2,2,i);
pcolor(lon-360,lat,ssh(:,:,ind(i)));
shading interp,colormap jet;
xlim([-77.2500  -71.9014]);,ylim([-38.0468  -33.4471])
caxis([0.35 0.65]),title(num2str(idx(ind(i),:)))
hold on,plot(-73.1217,-36.4137,'ok','MarkerFaceColor','k');
fillseg(mapa(70000:120000,:),[.5 .5 .5],[.5 .5 .5]);
end;

subplot(2,2,3);
ylabel('Latitude','Interpreter','latex');,xlabel('Longitude','Interpreter','latex');
colorbar('west');
packfig(2,2);
```

```matlab
[longitude latitude] = meshgrid(lon,lat);
x = [1 19];,y = [28 13];
c2 = improfile(longitude,x,y);
for i = 1:4;
c1 = improfile(ssh(:,:,ind(i)),x,y);	
hold on,plot(c2-360,vfilt(c1,3,'mirror')); 
end,clear c1;
axis tight,box on;
c_refline(-73.1217,'.-r','y');
ylabel('SSH(m)','Interpreter','latex');
xlabel('Longitude','Interpreter','latex');
legend('1','2','3','4','5','Interpreter','latex');	
title('January','Interpreter','latex');
```
